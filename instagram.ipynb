```python
import pandas as pd
import numpy as np

# ë¼ì´ë¸ŒëŸ¬ë¦¬ import
# ë¼ì´ë¸ŒëŸ¬ë¦¬ : í•„ìš”í•œ ë„êµ¬
from selenium import webdriver  # ë¼ì´ë¸ŒëŸ¬ë¦¬(ëª¨ë“ˆ) ê°€ì ¸ì˜¤ë¼
from selenium.webdriver import ActionChains as AC
import chromedriver_autoinstaller
from tqdm import tqdm
from tqdm import tqdm_notebook
import re
from time import sleep
import time

# ì›Œë‹ ë¬´ì‹œ
import warnings
warnings.filterwarnings('ignore')
```

```python
# ë°ì´í„° ìˆ˜ì§‘í•  í‚¤ì›Œë“œ ì§€ì •
keyword = "ì¹´í˜"
len_insta = 20
```

```python
# í¬ë¡¬ì°½ ë„ìš°ê¸°
chrome_path = chromedriver_autoinstaller.install()
driver = webdriver.Chrome(chrome_path)

driver.get("https://www.instagram.com/")
time.sleep(3)
```

```python
# ê²€ìƒ‰ì°½ì— ì»¤ì„œ í´ë¦­
element = driver.find_element_by_css_selector(".TqC_a")
element.click()
time.sleep(1)

# ê²€ìƒ‰ ì°½ì— ê²€ìƒ‰ì–´ ì…ë ¥
element = driver.find_element_by_css_selector(".XTCLo.Ju1dg.x3qfX")
element.clear()  # í˜¹ì‹œ ê²€ìƒ‰ì°½ì— ì¡´ì¬í•˜ëŠ” í…ìŠ¤íŠ¸ ì œê±°
element.send_keys(keyword)
time.sleep(2)

# ê²€ìƒ‰ì–´ ë¦¬ìŠ¤íŠ¸ ì¤‘ ì²«ë²ˆì§¸ ê²€ìƒ‰ì–´ í´ë¦­
query_list = driver.find_elements_by_css_selector(".-qQT3")
query_list[1].click()
time.sleep(3)
```

```python
dict = {}  # ì „ì²´ ê²Œì‹œê¸€ì„ ë‹´ì„ ê·¸ë¦‡

# ì²«ë²ˆì§¸ ì‚¬ì§„ í´ë¦­
CSS_tran="div.Nnq7C.weEfm .eLAPa"                         
driver.find_element_by_css_selector(CSS_tran).click()   # ì‚¬ì§„ í´ë¦­
time.sleep(2)

# í¬ë¡¤ë§ ì‹œì‘
for j in tqdm_notebook(range(0, len_insta)):    

    target_info = {}                                            # ì‚¬ì§„ë³„ ë°ì´í„°ë¥¼ ë‹´ì„ ë”•ì…”ë„ˆë¦¬ ìƒì„±

    try:    # í¬ë¡¤ë§ì„ ì‹œë„í•´ë¼.
        # ì‚¬ì§„(pic) í¬ë¡¤ë§
        overlays1 = "div._2dDPU.CkGkG .FFVAD"                   # ì‚¬ì§„ì°½ ì† ì‚¬ì§„   
        img = driver.find_element_by_css_selector(overlays1)    # ì‚¬ì§„ ì„ íƒ
        pic = img.get_attribute('src')                          # ì‚¬ì§„ url í¬ë¡¤ë§ ì™„ë£Œ
        target_info['picture'] = pic

        # ë‚ ì§œ(date) í¬ë¡¤ë§
        overlays2 = "div._2dDPU.CkGkG .c-Yi7 > time"                # ë‚ ì§œ ì§€ì •
        datum2 = driver.find_element_by_css_selector(overlays2)     # ë‚ ì§œ ì„ íƒ
        date = datum2.get_attribute('title')
        target_info['date'] = date

        # ì¢‹ì•„ìš”(like) í¬ë¡¤ë§
        overlays3 = ".Nm9Fw"                                        # ë¦¬ë·°ì°½ ì† ë‚ ì§œ
        datum3 = driver.find_element_by_css_selector(overlays3)     # ë¦¬ë·° ì„ íƒ
        like = datum3.text                                          # ì¢‹ì•„ìš” í¬ë¡¤ë§ ì™„ë£Œ
        target_info['like'] = like

        # í•´ì‹œíƒœê·¸(tag) í¬ë¡¤ë§
        overlays4 = ".xil3i"                                         
        datum3 = driver.find_elements_by_css_selector(overlays4)    # íƒœê·¸ ì„ íƒ
        tag_list = []
        for i in range(len(datum3)):
            tag_list.append(datum3[i].text)
        target_info['tag'] = tag_list

        dict[j] = target_info            # í† íƒˆ ë”•ì…”ë„ˆë¦¬ë¡œ ë§Œë“¤ê¸°

        print(j, tag_list)

        # ë‹¤ìŒì¥ í´ë¦­
        CSS_tran2="body > div._2dDPU.QPGbb.CkGkG > div.EfHg9 > div > div > div.l8mY4.feth3 > button"             # ë‹¤ìŒ ë²„íŠ¼ ì •ì˜
        tran_button2 = driver.find_element_by_css_selector(CSS_tran2)  # ë‹¤ìŒ ë²„íŠ¼ find
        AC(driver).move_to_element(tran_button2).click().perform()     # ë‹¤ìŒ ë²„íŠ¼ í´ë¦­
        time.sleep(3)

    except:  # ì—ëŸ¬ê°€ ë‚˜ë©´ ë‹¤ìŒì¥ì„ í´ë¦­í•´ë¼
        # ë‹¤ìŒì¥ í´ë¦­
        CSS_tran2="body > div._2dDPU.QPGbb.CkGkG > div.EfHg9 > div > div > div.l8mY4.feth3 > button"             # ë‹¤ìŒ ë²„íŠ¼ ì •ì˜
        tran_button2 = driver.find_element_by_css_selector(CSS_tran2)  # ë‹¤ìŒ ë²„íŠ¼ find
        AC(driver).move_to_element(tran_button2).click().perform()     # ë‹¤ìŒ ë²„íŠ¼ í´ë¦­

        time.sleep(3)

print(dict)
```
